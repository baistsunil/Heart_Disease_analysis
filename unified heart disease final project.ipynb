# Draw correlation matrix

```python
plt.figure(figsize=(12,6))
sns.heatmap(data.corr(),annot = True)
plt.show()
```

```python
# 'cp' , 'thaslach' , 'slope' are showing  good  postive correlation  to target 
```

# Heart disease rates

```python
data['target'].value_counts()
```

```python
# 0 - no.of people doesn't cotain heart disease
# 1 - no.of people cotain heart disease
plt.figure(figsize=(4, 5))
sns.countplot(data['target'])
plt.xticks([0,1],['No-Disease','Disease'])
plt.show()
```

# Heart rate cause by gender

```python
# count of gender

data['sex'].value_counts()
```

```python
sns.countplot(x='sex',data=data,hue='target')
plt.title('heart disease cause by gender')
plt.xticks([0,1],['female','male'])
plt.legend(labels=['No-Disease','Disease'])
plt.show()
```

# Heart disease cause by age

```python
sns.distplot(data['age'],bins=20)
plt.show()
```

# chest pain as per our target

```python
sns.countplot(x='cp',data=data,hue='target')
plt.legend(labels=['No-Disease','Disease'])
plt.show()
```

# Fasting blood sugar as per our target

```python
# with fbs > 120 is consider diabatics 
sns.countplot(x='fbs',data=data,hue='target')
plt.xlabel('Fasting blood sugar')
plt.legend(labels=['No-Disease','Disease'])
plt.show()
```

# Chest pain types

```python
# Typical Angina (TA), 
# Atypical Angina (ATA), 
# Non-Anginal Pain (NAP)
# Asymptomatic (ASY)
```

```python
sns.countplot(data['cp'])
plt.xlabel('Chest Pain')
plt.xticks([0,1,2,3],['TA','ATA','NAP','ASY'])
plt.show()
```
